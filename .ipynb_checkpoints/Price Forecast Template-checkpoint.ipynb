```python
from fbprophet import Prophet
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/example_wp_log_peyton_manning.csv')
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```
